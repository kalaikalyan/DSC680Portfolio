```python
#Load necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Function to read a csv file is read_csv
df = pd.read_csv("sales_record.csv")
df
```

```python
df_subset = df[[ 'Region', 'Unit Price','Total Profit','Item Type']]
df_subset.head()
```

```python
df1 = df_subset.groupby('Region').describe()['Total Profit']
df1
```

```python
#Scatter plot
plt.xlabel('Total Profit',fontsize = 25)
plt.ylabel('Item Type',fontsize = 25)
plt.scatter(df['Total Profit'],df['Item Type'])
```

```python
#  bar chart


plt.bar( df['Item Type'],df['Total Profit'])
plt.title('Total Profit',fontsize = 20)
plt.ylabel('Total Profit',fontsize = 15)
plt.xlabel('Item Type',fontsize = 15)
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='large'  
)
plt.grid(False)

plt.show()
```

```python
#Line chart
plt.plot( df1['count'],df1['mean'])

plt.title('Total Profit',fontsize = 20)
plt.xlabel('count',fontsize = 15)
plt.ylabel('mean',fontsize = 15)
plt.grid(True)
plt.show()
```

```python
#Histogram
plt.ylabel('Frequency')
plt.xlabel('Total Profit')
df["Item Type"],df['Total Profit'].hist(bins=20)
```
